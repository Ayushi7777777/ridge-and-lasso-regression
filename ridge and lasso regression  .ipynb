# Independent and Dependent Features

```python
from sklearn.model_selection import train_test_split
```

```python
x= dataset # Independent
y= df.target # Dependent
```

```python
x_train , y_train ,x_test , y_test = train_test_split(x,y , test_size=0.3, random_state=42)
```

```python
x_train
```

# Standerdization

```python
from sklearn.preprocessing import StandardScaler
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=30, random_state=42)
```

```python
scalar = StandardScaler()
```

```python
x_train = scalar.fit_transform(x_train)
```

```python
x_test = scalar.transform(x_test)
```

# Ridge

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regression = Ridge()
```

```python
ridge_regression
```

```python
parameters_ridge = {'alpha':[1,2,4,5,7,9,10,12,13,18,20,30,40,50,60,70,80,90]}
ridgecv = GridSearchCV(ridge_regression, parameters_ridge,scoring='neg_mean_squared_error',cv=5)
ridgecv.fit(x_train,y_train)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridge_predict = ridgecv.predict(x_test)
```

```python
ridge_predict
```

```python
sns.displot(ridge_predict-y_test ,kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ridge_predict,y_test)
```

```python
score
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
```

```python
parameters_lasso = {'alpha':[1,2,4,5,7,9,10,12,13,18,20,30,40,50,60,70,80,90]}
lassocv = GridSearchCV(lasso, parameters_lasso , scoring= 'neg_mean_squared_error', cv=5)
lassocv.fit(x_train,y_train)
```

```python
print(lassocv.best_params_)
```

```python
print(lassocv.best_score_)
```

```python
lasso_predict= lassocv.predict(x_test)
```

```python
lasso_predict
```
